```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_function(x, mu, sigma):
    return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Örnek kullanım
x_values = np.linspace(-5, 5, 100)
mu = 0
sigma = 1

y_values = gaussian_function(x_values, mu, sigma)

plt.plot(x_values, y_values, label=f'Gaussian, μ={mu}, σ={sigma}')
plt.title('Gaussian Fonksiyonu')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

#Bu kod, belirli bir μ (ortalama) ve σ (standart sapma) değeri için Gaussian fonksiyonunu hesaplar ve grafiğini çizer.
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def gaussian_filter(image, kernel_size, sigma):
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)

# Örnek kullanım
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)  # Görüntüyü gri tonlamalı olarak oku
kernel_size = 5
sigma = 1.0

smoothed_image = gaussian_filter(image, kernel_size, sigma)

# Görüntüleri yan yana gösterme
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Orijinal Görüntü')

plt.subplot(1, 2, 2)
plt.imshow(smoothed_image, cmap='gray')
plt.title('Gaussian Filtreleme Sonrası')

plt.show()
```

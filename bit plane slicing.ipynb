```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/kaggle/input/bit-plane-slicing/Ekran grnts 2023-12-19 190309.png', cv2.IMREAD_GRAYSCALE)

# Görüntüyü 8 bitlik ikili forma dönüştür
_, binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# Bit plane slicing fonksiyonu
def bit_plane_slice(image, bit):
    return (image >> bit) & 1

# Görüntü boyutlarını al
height, width = binary_image.shape

# Bir pencere oluştur
plt.figure(figsize=(15, 8))

# Orjinal görüntüyü göster
plt.subplot(3, 3, 1)
plt.imshow(binary_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Her bir bit düzeyini göster
for bit in range(8):
    bit_plane = bit_plane_slice(binary_image, bit)
    plt.subplot(3, 3, bit + 2)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'Bit {bit}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/kaggle/input/bit-plane-slicing/Ekran grnts 2023-12-19 190309.png', cv2.IMREAD_GRAYSCALE)

_, binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

def bit_plane_slice(image, bit):
    return (image >> bit) & 1

num_bits = 8
fig, axs = plt.subplots(1, num_bits, figsize=(15, 3))

for bit in range(num_bits):
    bit_plane = bit_plane_slice(binary_image, bit)
    axs[bit].imshow(bit_plane, cmap='gray')
    axs[bit].set_title(f'Bit {bit}')
    axs[bit].axis('off')

plt.show()
```
